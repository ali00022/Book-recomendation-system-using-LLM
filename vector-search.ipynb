```python
import pandas as pd
books = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Book recommendation system using llm/books_cleaned.csv')
books
```

```python
books["tagged_description"].to_csv("tagged_descriptions.txt", sep="\n", index=False, header=False)
```

```python
raw_documents = TextLoader("tagged_descriptions.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=embedding_model
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
import re

def retrieve_semantic_recommendations(query: str, top_k: int = 10) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k=50)

    books_list = []

    for rec in recs:
        first_word = rec.page_content.strip('"').split()[0]  # Get the first part
        isbn_match = re.match(r'^\d+', first_word)  # Extract only numbers

        if isbn_match:  # If a valid ISBN is found
            books_list.append(int(isbn_match.group()))  # Convert and add to list

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
